```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Range1d
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
output_notebook()

##############################################################

def get_positions(car_history):
    
  xs, ys, thetas = [], [], []
    
  for car in car_history:
    xs.append(car.x)
    ys.append(car.y)
    thetas.append(car.theta + pi + pi / 4)
    
  return xs, ys, thetas

##############################################################
r = 25.
my_position = {'x': 0., 'y': r,'theta': pi / 2.}
my_settings = {'tolerance': 0.001, 
                'length': 20.,
                'max_steering': pi / 4.,
                'steering_drift': 0.,
                'steering_noise': pi / 100.,
                'speed_noise': 0.01
                }

car_history, error_history = follow_race_track(tp = 10. , ti = 0.0 , td = 15.0 , n = 300, 
                                 car_settings = my_settings, initial_position = my_position)

xs, ys, thetas = get_positions(car_history)

##############################################################

source = ColumnDataSource(data = {
    'x' : xs,
    'y' : ys, 
    'theta' : thetas
})

p = figure(plot_width = 1000, plot_height = 500)

plot_car = p.triangle(
    'x', 'y', 
    size = 15, 
    fill_color = "violet",
    line_color = "violet",
    fill_alpha = 0.10,
    line_width = 1,
    angle = 'theta',
    legend = "car", 
    source = source)

p.circle(
  x = 75, 
  y = 25, 
  radius = 25,
  fill_color = "orange",
  line_color = "white",
  fill_alpha = 0.10,
  line_width = 0.,    
)

p.circle(
  x = 25, 
  y = 25, 
  radius = 25,
  fill_color = "orange",
  line_color = "white",
  fill_alpha = 0.10,
  line_width = 0.,    
)

p.square(
  x = 50, 
  y = 25, 
  size = 315,
  fill_color = "yellow",
  line_color = "white",
  fill_alpha = 0.15,
  line_width = 0.,    
)

p.x_range = Range1d(-25, 125)
p.y_range = Range1d(-10, 60)

##############################################################

def update(p, i, d):
    
  my_position = {'x': 0., 'y': 25., 'theta': pi / 2.}
  my_settings = {'tolerance': 0.001, 
                'length': 20.,
                'max_steering': pi / 4.,
                'steering_drift': 0.,
                'steering_noise': pi / 100.,
                'speed_noise': 0.01
                }

  car_history, error_history = follow_race_track(tp = p , ti = i , td = d , n = 300, 
                                 car_settings = my_settings, initial_position = my_position)

  xs, ys, thetas = get_positions(car_history)
  plot_car.data_source.data['x'] = xs
  plot_car.data_source.data['y'] = ys
  plot_car.data_source.data['theta'] = thetas
  
  print("tp:", p)
  print("ti:", i)
  print("td:", d)
  print("last error:", error_history[-1])
  print("average error of last 100 steps:", sum(error_history[-100:]) / 100)
  print ("average error", sum(error_history) / len(error_history))
  push_notebook()
    
##############################################################

show(p, notebook_handle = True)
interact(update, p = (0., 20., 1.), i = (-1., 1., 0.1) , d = (0, 30., 1.))

#
```
